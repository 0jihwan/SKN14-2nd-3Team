### 컬럼 정의서

- CustomerID : 고유한 회원ID (int)
- Churn : 이탈 여부 (0/1, int)
- Tenure : 고객의 e커머스 사용 기간 (, float)
- PreferredLoginDevice : 고객의 선호 접속 기기 (object)
- CityTier : 시티 티어(대도시/중도시/소도시를 미국에선 cityTier라 하는듯) (int)
- WarehouseToHome : 배송시간    (float)
- PreferredPaymentMode : 결제 수단  (object)
- Gender : 고객 성별 (MALE/FEMALE, object)
- HourSpendOnApp : 모바일 어플 사용시간      (float)
- NumberOfDeviceRegistered : 로그인된 기기 숫자     (int)
- PreferedOrderCat : 저번 달 고객이 선호했던 카테고리     (object)
- SatisfactionScore : 고객이 평가한 만족도 (1~5점, int)
- MaritalStatus : 혼인 여부     (Married/Single/Devorced, object)
- NumberOfAddress : 고객이 등록한 주소의 개수      (int)
- Complain : 지난 달에 컴플레인 제기 여부 (0/1, int)
- OrderAmountHikeFromlastYear : 지난 해 대비 상승한 주문 수(퍼센트)   (float)
- CouponUsed : 지난 달 사용 쿠폰       (자료형은 float, 수치는 int)
- OrderCount : 지난 달 주문 건수       (자료형은 float, 수치는 int)
- DaySinceLastOrder : 마지막 주문 날짜로부터 며칠이 지났는지 (자료형은 float, 수치는 int)
- CashbackAmount : 지난 달 캐시백 금액(ex. 구매 금액 중 1% 캐시백 ...)  (float)
